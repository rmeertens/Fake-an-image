```python

from tensorflow.python.framework.ops import reset_default_graph
reset_default_graph()
g = tf.get_default_graph()

input_placeholder = tf.placeholder(tf.float32, [1, 224,224,3])
input_wanted = tf.placeholder(tf.int32,[1])
to_add_image = tf.Variable(tf.random_normal([224,224,3], mean=0.0, stddev=0.1, dtype=tf.float32))

combined_images_not_clamped = input_placeholder+to_add_image
filledmax = tf.fill(tf.shape(combined_images_not_clamped), 1.0)
filledmin = tf.fill(tf.shape(combined_images_not_clamped), 0.0)
greater_than_one = tf.greater(combined_images_not_clamped, filledmax)

combined_images_with_max = tf.where(greater_than_one, filledmax, combined_images_not_clamped)
lower_than_zero =tf.less(combined_images_with_max, filledmin)
combined_images = tf.where(lower_than_zero, filledmin, combined_images_with_max)

net = vgg16.get_vgg_model()
tf.import_graph_def(net['graph_def'], name='vgg')

style_layer = 'prob:0'
the_prediction = tf.import_graph_def(
    net['graph_def'],
    name='vgg',
    input_map={'images:0': combined_images},return_elements=[style_layer])


goldfish_expected_tf = tf.one_hot(input_wanted, 1000, dtype=tf.float32, name="one-hot-preferred")
loss = tf.reduce_sum(tf.square(the_prediction[0]-goldfish_expected_tf))

optimizer = tf.train.AdamOptimizer().minimize(loss,var_list=[to_add_image])


print(the_prediction)
```

```python
sess = tf.InteractiveSession()

```

```python
def show_many_images(*images):
    fig = plt.figure()
    for i in range(len(images)):
        print(images[i].shape)
        subplot_number = 100+10*len(images)+(i+1)
        plt.subplot(subplot_number)
        plt.imshow(images[i])
    plt.show()
def load_image(path,imagename):
    the_path = os.path.join(path,imagename)
    print(the_path)
    pandaimage = plt.imread(the_path)
    pandaimage = vgg16.preprocess(pandaimage)
    return np.array([pandaimage])
```

```python
import random
with open('trainlabels.txt','a') as output_file:
    for index,image_name in enumerate(images_in_folder[380:]):
        try:
            print(str(index) + " of " + str(len(images_in_folder)))
            img_4d = load_image(real_image_path,image_name)

            sess.run(tf.global_variables_initializer())
            predicted_class = 10001
            to_predict_class = random.randint(0,1)
            while predicted_class != to_predict_class:
                _, whoooutput,pred,concat = sess.run([optimizer,loss,the_prediction,combined_images],feed_dict={input_placeholder:img_4d,input_wanted:[to_predict_class]})  
                predicted_class = np.argmax(pred)
            #show_many_images(img_4d[0],concat[0])

            to_use_filename = image_name.split(".")[0]
            to_use_filename += ".npy"
            np.save(to_use_filename, concat[0])
            a = np.load(to_use_filename)
            output_file.write(to_use_filename+"," + str(to_predict_class)+"\n")
            #print((a==concat[0]).all())
        except Exception as e:
            print("exception!")
```
